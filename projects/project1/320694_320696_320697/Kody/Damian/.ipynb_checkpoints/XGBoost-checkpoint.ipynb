### Load Data

```python
import openml
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# DATASETS IDS:
# 37 - diabeties
# 31 - creditg
# 44 - spambase
# 40597 - 40597

data_id = 40597
# List all datasets and their properties
openml.datasets.list_datasets(output_format="dataframe")

# Get dataset by ID
diabeties = openml.datasets.get_dataset(data_id)

X, _, _, _ = diabeties.get_data(dataset_format="dataframe")
# print(X)
#y = X.loc[:, 'class']
y = X.loc[:, 'Class1']
#X = X.drop(['class'], axis = 1)
X = X.drop(X.filter(regex='Class').columns, axis=1)
print('Dane X', X.head())
print('Target: ', y.head())



missing_values_per_column = X.isna().sum()
print("Brakujące wartości w każdej kolumnie:\n", missing_values_per_column)

missing_rows_count = X.isna().any(axis=1).sum()
print("Liczba wierszy z przynajmniej jedną brakującą wartością:", missing_rows_count)


label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
print(y[:5])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
print(len(X_train))
```

```python
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
from sklearn.compose import ColumnTransformer, make_column_selector
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier, XGBRegressor
from sklearn.model_selection import GridSearchCV
### operacje dla kolumn numerycznych
num_pipeline = Pipeline(steps=[
    ('impute',SimpleImputer(strategy='mean')),
    ('scale',MinMaxScaler())
])

### operacje dla kolumn kategorycznych
cat_pipeline = Pipeline(steps=[
    ('impute', SimpleImputer(strategy='most_frequent')),
    ('one-hot',OneHotEncoder(handle_unknown='ignore', sparse_output=False))

])

#### W funkcji ColumnTransformer podajemy liste transformerow dla poszczegolnych zestawow kolumn- (name, transformer, columns)
#### UWAGA: zamiast column selector (make_column_selector( dtype_include= np.number)) mozna podac wektor nazw kolumn, ale to rozwiazanie dla konkretnych danych


col_trans = ColumnTransformer(transformers=[
    ('num_pipeline', num_pipeline, make_column_selector( dtype_include= np.number)),
    ('cat_pipeline',cat_pipeline,make_column_selector( dtype_include='category'))
    ],
    remainder='passthrough',
    n_jobs=-1)


model_pipeline = Pipeline([('preprocessing', col_trans),
                           ('model', XGBClassifier())])
# param_grid = {
#     'model__n_estimators': [100, 200, 300],
#     'model__learning_rate': [0.01, 0.1, 0.2],
#     'model__max_depth': [3, 5, 7],
#     'model__subsample': [0.8, 1.0],
#     'model__colsample_bytree': [0.8, 1.0],
#     'model__gamma': [0, 1, 5],
#     'model__min_child_weight': [1, 3, 5]
# }

# # Tuning hiperparametrów z GridSearchCV
# grid_search = GridSearchCV(model_pipeline, param_grid, cv=5, n_jobs=-1, verbose=1)

# # Fitting modelu z tuningiem
# grid_search.fit(X_train, y_train)

# # Najlepsze hiperparametry
# print("Best parameters found: ", grid_search.best_params_)

# # Można uzyskać najlepszy model
# best_model = grid_search.best_estimator_

# # Można sprawdzić jego dokładność
# best_model_score = best_model.score(X_test, y_test)
# print("Best model accuracy on test data: ", best_model_score)
```

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform, randint
import numpy as np
from sklearn.model_selection import RandomizedSearchCV, StratifiedKFold
from sklearn.metrics import roc_auc_score, r2_score
import pandas as pd

param_distributions = {
    'model__n_estimators': randint(100, 1000),
    'model__learning_rate': uniform(0.01, 0.2),
    'model__max_depth': randint(1, 10),
    'model__subsample': uniform(0.1, 0.9),
    'model__min_child_weight': uniform(0, 7),
    'model__colsample_bytree': uniform(0.1, 0.9),
    'model__colsample_bylevel': uniform(0.1, 0.9),
}

n_iter_values = [250]
scores = []

cv_strategy = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for n_iter in n_iter_values:
    random_search = RandomizedSearchCV(
        model_pipeline,
        param_distributions=param_distributions,
        n_iter=n_iter,
        cv=cv_strategy,
        n_jobs=-1,
        verbose=1,
        random_state=42,
        scoring='roc_auc',
    )

    random_search.fit(X_train, y_train)

    best_model = random_search.best_estimator_
    best_model_score = best_model.score(X_test, y_test)
    scores.append(best_model_score)
    score = random_search.best_score_
    print("Score from randomsearch", score)
    print(f"Iterations: {n_iter}, Best Accuracy: {best_model_score:.4f}")
    results = random_search.cv_results_

    results_df = pd.DataFrame(random_search.cv_results_)
    
    results_df.to_csv(f'random_search_results_data{data_id}.csv')


```

```python
# Wczytanie wyników z pliku CSV do DataFrame
# DATASETS IDS:
# 37 - diabeties
# 31 - creditg
# 44 - spambase
# 40597 - 40597
results_37 = pd.read_csv('random_search_results_data37.csv')
results_31 = pd.read_csv('random_search_results_data31.csv')
results_44 = pd.read_csv('random_search_results_data44.csv')
results_40597 = pd.read_csv('random_search_results_data40597.csv')

# Wczytanie kolumny 'mean_test_score' do zmiennej
test_score_37 = results_37['mean_test_score']
test_score_31 = results_31['mean_test_score']
test_score_44 = results_44['mean_test_score']
test_score_40597 = results_40597['mean_test_score']

max_score = max(test_score_37)
min_score = min(test_score_37)

mean_test_score_37 = [(score ) / (max_score - min_score) for score in test_score_37]

max_score = max(test_score_31)
min_score = min(test_score_31)

mean_test_score_31 = [(score ) / (max_score - min_score) for score in test_score_31]

max_score = max(test_score_44)
min_score = min(test_score_44)

mean_test_score_44 = [(score ) / (max_score - min_score) for score in test_score_44]


max_score = max(test_score_40597)
min_score = min(test_score_40597)

mean_test_score_40597 = [(score ) / (max_score - min_score) for score in test_score_40597]


scores_df = pd.DataFrame({
    'mean_test_score_37': mean_test_score_37,
    'mean_test_score_31': mean_test_score_31,
    'mean_test_score_44': mean_test_score_44,
    'mean_test_score_40597': mean_test_score_40597
})
scores_df['mean_score_across_datasets'] = scores_df.mean(axis=1)

best_index = scores_df['mean_score_across_datasets'].idxmax()

best_score = scores_df.loc[best_index, 'mean_score_across_datasets']
print(f"Best average score across datasets: {best_score:.4f}")

# Wyciągamy hiperparametry dla najlepszej konfiguracji na podstawie indeksu
print("Best index: ", best_index)
best_params = results_37.loc[best_index]
print("Best hyperparameters:")
print(best_params)

best_params = results_37.loc[best_index, results_37.columns.str.startswith('param_model__')]

print(test_score_37[best_index])
print(test_score_31[best_index])
print(test_score_44[best_index])
print(test_score_40597[best_index])
# Zapisujemy hiperparametry do pliku CSV
best_params.to_frame().transpose().to_csv('optimal_xgboost_params.csv', index=False)
```

### Caluclate hyperparameter tunability

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform, randint
import numpy as np
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import roc_auc_score, r2_score
import pandas as pd

# Parametry do RandomizedSearchCV
param_distributions = {
    'model__n_estimators': randint(100, 1000),
    'model__learning_rate': uniform(0.01, 0.2),
    'model__max_depth': randint(1, 10),
    'model__subsample': uniform(0.1, 0.9),
    'model__min_child_weight': uniform(0, 7),
    'model__colsample_bytree': uniform(0.1, 0.9),
    'model__colsample_bylevel': uniform(0.1, 0.9),
}


def remove_param_prefix(params_dict):
    return {key.replace('param_', ''): value for key, value in params_dict.items()}
# Lista liczby iteracji do przetestowania
n_iter_values = [50]
scores = []

cv_strategy = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Wczytanie najlepszych hiperparametrów z pliku
best_params_df = pd.read_csv('optimal_xgboost_params.csv')
best_params = best_params_df.to_dict(orient='records')[0]
best_params = remove_param_prefix(best_params)

# Pętla po parametrach, aby testować tunowalność tylko jednego parametru
for key, value in param_distributions.items():
    print(f"Sprawdzam tunowalność parametru {key} z wartością {value}")
    
    # Kopiowanie najlepszych parametrów i zmiana tylko jednego
    current_best_params = best_params.copy()

    # Przygotowanie modelu z najlepszymi parametrami
    model_pipeline.set_params(**current_best_params)

    # RandomizedSearchCV z różną liczbą iteracji
    random_search = RandomizedSearchCV(
        model_pipeline,
        param_distributions={key: value},  # Przekazujemy tylko aktualnie testowany parametr
        n_iter=n_iter_values[0],
        cv=cv_strategy,
        n_jobs=-1,
        verbose=1,
        random_state=42,
        scoring='roc_auc',
    )

    # Fitowanie modelu
    random_search.fit(X_train, y_train)

    # Pobieranie najlepszego modelu i wyniku
    best_model = random_search.best_estimator_
    score = random_search.best_score_

    print(f"Score z RandomizedSearch: {score:.4f}")
    print(f"Iterations: {n_iter_values[0]}, Best AUC: {score:.4f}")

    # zamien test_score_40597 w zaleznosci od datasetu
    tunability = score - test_score_40597[best_index]
    print(f"Tunability of {key}: {tunability:.4f}")
    # Przekształcenie wyników do DataFrame
    results_df = pd.DataFrame(random_search.cv_results_)
    
    # Zapisanie DataFrame do pliku CSV
    results_df.to_csv(f'{key}_random_search_results_data{data_id}.csv')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

def plot_best_mean_test_score(csv_file_path):
    # Wczytanie pliku CSV
    df = pd.read_csv(csv_file_path)

    # Załóżmy, że kolumna 'mean_test_score' to średni wynik testowy
    # Możemy obliczyć najlepszy wynik do każdej iteracji (przechodzimy przez wszystkie iteracje i bierzemy najlepszy wynik do tej chwili)
    df['best_mean_test_score'] = df['mean_test_score'].cummax()

    # Numer iteracji to numer wiersza (indeks + 1, ponieważ indeksy w pandas zaczynają się od 0)
    iterations = df.index + 1

    # Tworzymy wykres
    plt.figure(figsize=(10, 6))
    plt.plot(iterations, df['best_mean_test_score'], color='b', label='Best Mean Test Score')

    # Dodajemy etykiety osi
    plt.xlabel('Numer Iteracji')
    plt.ylabel('Najlepszy Średni Wynik Testowy')
    plt.title('Zależność najlepszego średniego wyniku testowego od numeru iteracji')

    # Dodajemy nazwę pliku na górze wykresu
    file_name = os.path.basename(csv_file_path)  # Pobranie nazwy pliku
    plt.suptitle(f'Wykres dla pliku: {file_name}', fontsize=12, y=1.05)

    # Wyświetlamy wykres
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
plot_best_mean_test_score('random_search_results_data37.csv')
plot_best_mean_test_score('random_search_results_data31.csv')
plot_best_mean_test_score('random_search_results_data44.csv')
plot_best_mean_test_score('random_search_results_data40597.csv')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

def plot_boxplot(dataframe, column_names):
    """
    Rysuje box plot dla wybranych kolumn w jednym wykresie, nadając każdej kolumnie automatycznie wygenerowany kolor.

    Args:
        dataframe (pd.DataFrame): DataFrame zawierający dane.
        column_names (list): Lista nazw kolumn do wyświetlenia.
    """
    if not all(col in dataframe.columns for col in column_names):
        raise ValueError("Jedna lub więcej kolumn nie istnieje w DataFrame.")
    
    long_data = dataframe[column_names].melt(var_name="Kolumna", value_name="Wartości")
    
    colors = sns.color_palette("husl", len(column_names))  # Husl to paleta kolorów
    palette = dict(zip(column_names, colors))

    sns.boxplot(x="Kolumna", y="Wartości", data=long_data, palette=palette)

    plt.title("Box Plot dla wybranych datasetów")
    plt.xlabel("Kolumny")
    plt.ylabel("Wartości")

    plt.show()
```
