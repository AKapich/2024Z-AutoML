```python
import matplotlib.pyplot as plt
import pandas as pd

test_type = ""
```

```python
fig_random, axes_random = plt.subplots(2, 2, figsize=(12, 10))
fig_random.suptitle("Convergence Plots of Random Search", fontsize=16)

for i in range(4):
    random_results = pd.read_csv(f'{test_type}_random_iter_res_{i}.csv')
    random_scores = random_results['mean_test_score']
    random_best_so_far = random_scores.cummax()
    
    ax = axes_random[i // 2, i % 2]
    ax.plot(random_best_so_far[1:], marker='o', color='blue', markersize=3)
    ax.set_xlabel("Iteration")
    ax.set_ylabel("Best score so far (AUC)")
    ax.set_title(f"Dataset {i}")
    ax.grid()

plt.tight_layout(rect=[0, 0, 1, 0.96])
```

```python
fig_bayes, axes_bayes = plt.subplots(2, 2, figsize=(12, 10))
fig_bayes.suptitle("Convergence Plots of Bayesian Search", fontsize=16)

for i in range(4):
    bayes_results = pd.read_csv(f'{test_type}_bayes_iter_res_{i}.csv')
    bayes_scores = bayes_results['mean_test_score']
    bayes_best_so_far = bayes_scores.cummax()
    
    ax = axes_bayes[i // 2, i % 2] 
    ax.plot(bayes_best_so_far[1:], marker='o', color='green', markersize=3)
    ax.set_xlabel("Iteration")
    ax.set_ylabel("Best score so far (AUC)")
    ax.set_title(f"Dataset {i}")
    ax.grid()

plt.tight_layout(rect=[0, 0, 1, 0.96])
```
